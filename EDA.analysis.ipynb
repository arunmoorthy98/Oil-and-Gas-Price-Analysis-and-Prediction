```python
# 3. Time Series Analysis
# Plotting closing prices and volume over time
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
plt.plot(df.index, df['Close'], label='Close Price')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.bar(df.index, df['Volume'], width=30, label='Volume', color='orange')
plt.title('Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
# Replace this line with the correct path to your CSV file if not already loaded
# df = pd.read_csv('your_file_path.csv')

# Select only numeric columns
numeric_columns = df.select_dtypes(include='number')

# Basic summary statistics
summary_stats = numeric_columns.describe(include='all')

# Pairplot for visualizing relationships
sns.pairplot(numeric_columns)
plt.show()
```

```python
for col in numeric_columns.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(numeric_columns[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Correlation heatmap
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
